Crear Diferentes Tipos de Arreglos

```python
# Crear un arreglo ordenado
arreglo_ordenado = np.arange(0, 10, 3)
print("Arreglo Ordenado:", arreglo_ordenado)

# Crear un arreglo de ceros
arreglo_ceros = np.zeros(10)
print("Arreglo de Ceros:", arreglo_ceros)

# Crear un arreglo de unos con dimensiones específicas
arreglo_unos = np.ones((3, 5))
print("Arreglo de Unos:", arreglo_unos)
```

3. Crear un Arreglo con Números Aleatorios

```python
# Crear un arreglo de números aleatorios
arreglo_aleatorio = np.random.randint(0, 100, 10)
print("Arreglo Aleatorio:", arreglo_aleatorio)

# Crear un arreglo de números aleatorios constantes
np.random.seed(101)
arreglo_aleatorio_const = np.random.randint(0, 100, 10)
print("Arreglo Aleatorio Constante:", arreglo_aleatorio_const)
```

4. Funciones en Arreglos

```python
# Cambiar la forma del arreglo
nuevo_arreglo = arreglo_aleatorio.reshape((2, 5))
print("Nuevo Arreglo:", nuevo_arreglo)

# Encontrar el valor máximo y su índice
valor_maximo = arreglo_aleatorio.max()
indice_maximo = arreglo_aleatorio.argmax()
print("Valor Máximo:", valor_maximo)
print("Índice del Valor Máximo:", indice_maximo)
```

5. Abrir una Imagen y Convertirla en Arreglo

```python
import cv2
from matplotlib import pyplot as plt
from PIL import Image

# Leer la imagen usando PIL
pic = Image.open('imagen.png')
print("Tipo de la Imagen:", type(pic))

# Convertir la imagen en un arreglo
pic_arr = np.asarray(pic)
print("Forma del Arreglo de la Imagen:", pic_arr.shape)

# Mostrar la imagen usando matplotlib
plt.imshow(pic_arr)
plt.show()
```

Mostrar Imagen


```python
# Mostrar la imagen
plt.imshow(pic_arr)
plt.axis('off')
plt.show()
```

6. Modificar Colores de una Imagen

```python
# Mostrar solo el canal rojo de la imagen
pic_red = pic_arr.copy()
plt.imshow(pic_red[:, :, 0], cmap='gray')
plt.show()

# Quitar los canales verde y azul
pic_red[:, :, 1] = 0
pic_red[:, :, 2] = 0
plt.imshow(pic_red)
plt.show()
```

Cambiar Color de Imagen con OpenCV


```python
# Leer la imagen usando OpenCV
img = cv2.imread('imagen.png')
print("Tipo de la Imagen:", type(img))

# Corregir los colores de la imagen
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img)
plt.show()

# Convertir la imagen a escala de grises
fix_gray = cv2.imread('imagen.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(fix_gray, cmap='gray')
plt.show()
```

Dibujar Figuras en la Imagen


```python
# Crear una imagen en negro
blank_img = np.zeros((512, 512, 3), dtype=np.int16)
plt.imshow(blank_img)
plt.show()

# Dibujar un rectángulo
cv2.rectangle(blank_img, pt1=(384, 50), pt2=(500, 150), color=(0, 255, 0), thickness=10)
plt.imshow(blank_img)
plt.show()

# Dibujar un círculo
cv2.circle(blank_img, center=(100, 100), radius=50, color=(0, 0, 255), thickness=5)
plt.imshow(blank_img)
plt.show()

# Dibujar una línea
cv2.line(blank_img, pt1=(0, 0), pt2=(512, 512), color=(0, 100, 200), thickness=5)
plt.imshow(blank_img)
plt.show()

# Escribir texto en la imagen
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img, text='Hola', org=(10, 500), fontFace=font, fontScale=4, color=(255, 255, 255), thickness=2, lineType=cv2.LINE_AA)
plt.imshow(blank_img)
plt.show()
```
